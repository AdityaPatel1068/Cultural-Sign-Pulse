## Refernece Sheet

```python
import os
import pandas as pd

# Define folder paths
dataset_folders = {
    "train": r"D:\ASL\Dataset\train",
    "test": r"D:\ASL\Dataset\test",
    "val": r"D:\ASL\Dataset\val"
}

# Initialize an empty list to store file details
data = []

# Loop through each dataset folder and collect file names
for dataset, folder_path in dataset_folders.items():
    for file_name in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, file_name)):
            # Extract the label by removing the file extension
            label = os.path.splitext(file_name)[0]
            
            # Append the dataset name, file name, label, and an empty sentiment field to the data list
            data.append({
                "dataset": dataset,
                "file_name": file_name,
                "label": label,
                "sentiment": ""
            })

# Convert the list to a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
reference_csv_path = r"D:\ASL\Dataset\reference.csv"
df.to_csv(reference_csv_path, index=False)

print(f"Reference CSV file created at {reference_csv_path}")
```

# Sentiment Analysis


```python
import os
import pandas as pd
import cv2
from fer import FER

# Define paths using WSL format
dataset_folders = {
    "train": "/mnt/d/ASL/Dataset/train",
    "test": "/mnt/d/ASL/Dataset/test",
    "val": "/mnt/d/ASL/Dataset/val"
}
reference_csv_path = "/mnt/d/ASL/Dataset/reference.csv"

# Load the reference CSV
df = pd.read_csv(reference_csv_path)

# Initialize the FER model for sentiment analysis
emotion_detector = FER(mtcnn=True)

# Function to perform sentiment analysis on a video and get the dominant emotion
def analyze_sentiment(video_path):
    cap = cv2.VideoCapture(video_path)
    emotion_count = {}

    # Process frames for sentiment analysis
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Perform emotion detection on the frame
        emotions = emotion_detector.detect_emotions(frame)
        if emotions:
            # Get the most dominant emotion in the frame
            emotion_scores = emotions[0]["emotions"]
            dominant_emotion = max(emotion_scores, key=emotion_scores.get)

            # Update emotion count
            if dominant_emotion in emotion_count:
                emotion_count[dominant_emotion] += 1
            else:
                emotion_count[dominant_emotion] = 1

    # Release video capture object
    cap.release()

    # Determine the most frequently occurring emotion across frames
    if emotion_count:
        dominant_sentiment = max(emotion_count, key=emotion_count.get)
        return dominant_sentiment
    return "No face detected"  # Return if no emotion detected

# Update the DataFrame with the dominant sentiment for each video
for index, row in df.iterrows():
    dataset = row['dataset']
    file_name = row['file_name']
    
    # Determine the full path to the video file
    video_path = os.path.join(dataset_folders[dataset], file_name)
    
    if os.path.exists(video_path):
        # Perform sentiment analysis on the video
        dominant_sentiment = analyze_sentiment(video_path)
        
        # Update the 'sentiment' column in the DataFrame
        df.at[index, 'sentiment'] = dominant_sentiment
        print(f"Processed {file_name}: Dominant sentiment is {dominant_sentiment}")
    else:
        print(f"File {file_name} not found in {dataset} folder")

# Save the updated DataFrame back to the reference CSV
df.to_csv(reference_csv_path, index=False)

print("Sentiment analysis complete. Updated reference CSV file.")
```
